```python
import pandas as pd
```

```python
df = pd.read_csv('Spellman.csv')
```

```python
df.rename(columns={'time': 'gene'}, inplace=True)
```

```python
df = pd.melt(df, id_vars=['gene'], var_name = 'time', value_name = 'expression')
```

```python
df.to_csv('Spellman-tidy.csv', index=False)
```
